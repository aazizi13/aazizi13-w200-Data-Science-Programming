b) Below is a data frame of customers that have different cooling systems. Your data science team lead wants the column ```cooling_system``` to be labeled with the integer numbers 1-4 instead of the text as shown below:

- 1 = Air Conditioning / AC / Air Con
- 2 = Heat Pump / HP
- 3 = Evaporative Cooler / Evap Cooler
- 4 = Fan

Make a new column called ```cooling_type``` that maps the text values to the new numeric values. Filter out the values that are not included in the mapping above. Print out/display this new data frame. Be sure to list any assumptions also!

````python
import pandas

# creating a data frame from scratch - list of lists

data = [ [101, 'AC'], 
         [102, 'Heat Pump'], 
         [103, 'Air Con'], 
         [104, 'Air Conditioning'],
         [105, 'Fan'],
         [106, 'None'],
         [107, 'Evap Cooler'],
         [108, None],
         [109, 'AC'],
         [110, 'Evaporative Cooler'],
         [111, 'geothermal'],
         [112, 1]
       ]

# create a data frame with column names - list of lists

col_names = ['Cust_Number', 'Cooling_System']
df = pandas.DataFrame(data, columns=col_names)
df
````

````python
def finding_cooling_type (i):
    
    if i == 'Air Conditioning' or i == 'AC' or i == 'Air Con':
        return 1
    
    if i == 'Heat Pump' or i == 'HP':
        return 2
    
    if i == 'Evaporative Cooler' or i == 'Evap Cooler':
        return 3
    
    if i == 'Fan':
        return 4
    
df['cooling_type'] = df['Cooling_System'].apply(finding_cooling_type)
df.dropna(inplace = True)
df['cooling_type'] = df['cooling_type'].astype(int)

df
````

Also, visualize the record as two vertically arranged plots.

- The two plots' horizontal axes should align. Demonstrate that your code works by applying it to `data`.

Notes: <br>
- The gain for each period is the end balance minus the start balance. 
- The growth factor for each period is the end balance divided by the start balance.  
- The return for each period is the growth factor minus 1.

````python
data = ({"Jan 2018":1000},{"Feb 2018":1100},{"Mar 2018":1400},{"Apr 2018":700},{"May 2018":800},{"Jun 2018":500})
data
````

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

data = ({"Jan 2018":1000},{"Feb 2018":1100},{"Mar 2018":1400},{"Apr 2018":700},
        {"May 2018":800},{"Jun 2018":500})


month = []
balance = []
Return = []
last_balance = -1


for i in data:
    for key, val in i.items():
        month.append(key)
        balance.append(val)
        
        if last_balance != -1:
            R = (val/last_balance) - 1
            R = round(R, 2)
        else:
            R = 0.0
            
        last_balance = val
        Return.append(R)
        
my_data = pd.DataFrame([month, balance, Return]).T
my_data.columns = ['date', 'start_balance', 'return']
my_data
````

````python
color_coded = []
for i in Return:
    if i > 0:
        color_coded.append('g')
    else:
        color_coded.append('r')

fig, (ax1, ax2) = plt.subplots(2, sharex=True)
fig.suptitle('Bitcoin Return Visualization')

ax1.plot(my_data['date'], my_data['start_balance'])
ax1.set(ylabel="Start Balance")


ax2.bar(my_data['date'], my_data['return'], color=color_coded)
plt.axhline(y=0, color='black', linestyle='-')
plt.xticks(rotation=45) 
plt.xlabel("Month")
plt.ylabel("Return")
pass
````

### VCF file description (Summarized from version 4.1)

```
* The VCF specification:

VCF is a text file format which contains meta-information lines, a header line, and then data lines each containing information about a position in the genome. The format also can contain genotype information on samples for each position.

* Fixed fields:

There are 8 fixed fields per record. All data lines are **tab-delimited**. In all cases, missing values are specified with a dot (.). 

1. CHROM - chromosome number
2. POS - position DNA nuceleotide count (bases) along the chromosome
3. ID - The unique identifier for each mutation
4. REF - reference base(s)
5. ALT - alternate base(s)
6. FILTER - filter status
7. QUAL - quality
8. INFO - a semicolon-separated series of keys with values in the format: <key>=<data>

```
### Applicable INFO field specifications

```
GENEINFO = <Gene name>
CLNSIG =  <Clinical significance>
CLNDN = <Disease name>
```

````python
#reading the file and converting it into DataFrames. 

import io
import os
import numpy as np
import pandas as pd


def read_vcf(path):
    with open(path, 'r') as f:
        lines = [l for l in f if not l.startswith('#')]
    return pd.read_csv(
        io.StringIO(''.join(lines)),
        dtype={'#CHROM': str, 'POS': str, 'ID': str, 'REF': str, 'ALT': str,
               'QUAL': str, 'FILTER': str, 'INFO': str},
        sep='\t'
    ).rename(columns={'#CHROM': 'CHROM'})


df = read_vcf("clinvar_final.txt")
df.head()
````

````python
# Getting CLNDN from INFO column. 

def parse_CLNDN(info: str):

    split_data = info.split(";")
    for i in split_data:
        if "CLNDN" in i:
            return i.split("=")[-1]
````

````python
# Getting Gene info from INFO column. 

def parse_gene_info(info: str):
    
    split_data = info.split(";")
    for i in split_data:
        if "GENEINFO" in i:
            return i.split("=")[-1]
````

````python
# Getting CLNSIG from INFO column. 

def parse_CLNSIG(info: str):
    
    split_data = info.split(";")
    for i in split_data:
        if "CLNSIG" in i:
            return i.split("=")[-1]
````

````python
# creating new DataFrame with Required columns. 

new_df = []

for row in df.iterrows():
    info = row[1][-1]
    new_df.append(
        {
            "Gene_name": parse_gene_info(info),
            "Mutation_ID_number": row[1][2],
            "Mutation_Position": row[1][1],
            "Mutation_value": row[1][3] + row[1][4],
            "Clinical_Significance": parse_CLNSIG(info),
            "Disease_implicated": parse_CLNDN(info)
        }
    )

new_df = pd.DataFrame(new_df)
new_df.head(10)
````
